```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Geometric distribution.
geo = np.random.geometric(p=0.25, size=100)

mean = np.mean(geo)
print(mean)
std = np.std(geo)
print(std)

plt.hist(geo, bins=12, color='r')
plt.axvline(geo.mean(), color='g', linestyle='solid', linewidth=2)
plt.axvline(geo.mean() + geo.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(geo.mean() - geo.std(), color='g', linestyle='dashed', linewidth=2)
plt.show()
```

I find it very interesting that the Geometric Distribution has a standard deviation that goes outside the range of values!  I didn't realize that this was possible, but it makes sense if the majority of your values are close to one side of the range.  

```python
# Rayleigh distribution.
ray = np.random.rayleigh(scale=1.0, size=100)

mean = np.mean(ray)
print(mean)
std = np.std(ray)
print(std)

plt.hist(ray, bins=8, color='m')
plt.axvline(ray.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(ray.mean() + ray.std(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(ray.mean() - ray.std(), color='b', linestyle='dashed', linewidth=1)
plt.show()
```

The Rayleigh Distribution is much closer to what I would envision as a "normal" distribution.  Although the values are skewed to the left of the mean, it still gives the resemblance of a normal distribution. 

```python
# Wald distribution.
wald = np.random.wald(mean=2, scale=3, size=100)

mean = np.mean(wald)
print(mean)
std = np.std(wald)
print(std)

plt.hist(wald, bins=15, color='b')
plt.axvline(wald.mean(), color='y', linestyle='solid', linewidth=1.5)
plt.axvline(wald.mean() + wald.std(), color='y', linestyle='dashed', linewidth=1)
plt.axvline(wald.mean() - wald.std(), color='y', linestyle='dashed', linewidth=1)
plt.show()
```

The Wald Distribution reminds me of the geometric distribution with the parabolic structure and the standard deviation that is outside the low range.  

```python
# Exponential distribution.
exp = np.random.exponential(scale=1.0, size=100)

mean = np.mean(exp)
print(mean)
std = np.std(exp)
print(std)

plt.hist(exp, bins=10, color='g')
plt.axvline(exp.mean(), color='k', linestyle='solid', linewidth=3)
plt.axvline(exp.mean() + exp.std(), color='k', linestyle='dotted', linewidth=1.5)
plt.axvline(exp.mean() - exp.std(), color='k', linestyle='dotted', linewidth=1.5)
plt.show()
```

The Exponential Distribution also has the parabolic curve like the Geometric and Wald Distributions, which makes perfect sense mathematically.  

```python
# Laplace distribution.
lap = np.random.laplace(loc=0, scale=1.0, size=100)

mean = np.mean(lap)
print(mean)
std = np.std(lap)
print(std)

plt.hist(lap, bins=7, color='c')
plt.axvline(lap.mean(), color='b', linestyle='dotted', linewidth=2)
plt.axvline(lap.mean() + lap.std(), color='b', linestyle='solid', linewidth=2)
plt.axvline(lap.mean() - lap.std(), color='b', linestyle='solid', linewidth=2)
plt.show()
```

The Laplace Distribution is another that resembles the normal distribution.  It is interesting to note that all of the distributions have a size of 100, but some are covering a range of about 25 while others are only 3.5.  

```python
# Triangular distribution.
tri = np.random.triangular(left=-5.0, mode=3.0, right=13, size=100)

mean = np.mean(tri)
print(mean)
std = np.std(tri)
print(std)

plt.hist(tri, bins=12, color='y')
plt.axvline(tri.mean(), color='k', linestyle='dashed', linewidth=2)
plt.axvline(tri.mean() + tri.std(), color='k', linestyle='dotted', linewidth=1)
plt.axvline(tri.mean() - tri.std(), color='k', linestyle='dotted', linewidth=1)
plt.show()
```

The Triangular Distribution is also similar to the Laplace and Rayleigh Distributions, except it seems to have more values at the ranges, which is different from any other histogram.

```python
# Two normal distributions combined to make a third variable.

var1 = np.random.normal(loc=5, scale=0.5, size=100)
mean1 = np.mean(var1)
print(mean1)
std1 = np.std(var1)
print(std1)

var2 = np.random.normal(loc=10, scale=1, size=100)
mean2 = np.mean(var2)
print(mean2)
std2 = np.std(var2)
print(std2)

var3 = var1 + var2
mean3 = np.mean(var3)
print(mean3)
std3 = np.std(var3)
print(std3)

plt.hist(var3, bins=25, color='k')
plt.axvline(var3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The most interesting thing about this distribution is that mean1 + mean2 == mean3, but std1 + std2 !== std3.  It makes sense after thinking about the relationships, but it was not expected.
